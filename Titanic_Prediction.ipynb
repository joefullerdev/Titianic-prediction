```python
import pandas as pd
```

Handling train data

```python
trainDf = pd.read_csv('train.csv')
trainDf.head()
```

```python
newTrainDf = trainDf.drop(columns=['Name', 'Ticket', 'Cabin', 'Embarked'])
newTrainDf.head()
```

```python
newTrainDf['Sex'] = newTrainDf['Sex'].replace(['female'], 1)
newTrainDf.head()
```

```python
newTrainDf['Sex'] = newTrainDf['Sex'].replace(['male'], 2)
newTrainDf.head()
```

```python
newTrainDf.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(newTrainDf[['Age']])
newTrainDf['Age'] = imputer.fit_transform(newTrainDf[['Age']])
```

```python
newTrainDf.isnull().sum()
```

Handling Test Data

```python
testDf = pd.read_csv('test.csv')
testDf.head()
```

```python
newTestDf = testDf.drop(columns=['Name', 'Ticket', 'Cabin', 'Embarked'])
newTestDf.head()
```

```python
newTestDf['Sex'] = newTestDf['Sex'].replace(['female'], 1)
newTestDf.head()
```

```python
newTestDf['Sex'] = newTestDf['Sex'].replace(['male'], 2)
newTestDf.head()
```

```python
newTestDf.isnull().sum()
```

```python
imputer.fit(newTestDf[['Age']])
newTestDf['Age'] = imputer.fit_transform(newTestDf[['Age']])
```

```python
imputer.fit(newTestDf[['Fare']])
newTestDf['Fare'] = imputer.fit_transform(newTestDf[['Fare']])
```

```python
newTestDf.isnull().sum()
```

Building the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(multi_class='multinomial')
```

```python
X = newTrainDf.drop(columns=['Survived'])
y = newTrainDf['Survived']
```

```python
X.head()
```

```python
y.head()
```

```python
model.fit(X, y)
```

```python
Prediction = model.predict(newTestDf)
Prediction
```

```python
PredictionDf = pd.DataFrame(Prediction)
PredictionDf.head()
```

```python
PredictionDf['Survived'] = PredictionDf[0]
PredictionDf = PredictionDf.drop(columns=[0])
PredictionDf.head()
```

```python
PassID = newTestDf['PassengerId']
PassID.head()
```

```python
dfs = [PassID, PredictionDf]
result = pd.concat(dfs, axis=1)
result.head()
```

```python
result.info()
```

```python
result.to_csv('result.csv', index=False)
```
